```python
def index_documents(path: str | Path, save_to: str) -> GPTSimpleVectorIndex:
    documents = SimpleDirectoryReader(path).load_data()
    index = GPTSimpleVectorIndex(documents)
    index.save_to_disk(save_to)

    return index
```

```python
index = index_documents("project", "index_project.json")

# load from disk
# index = GPTSimpleVectorIndex.load_from_disk('index.json')
```
